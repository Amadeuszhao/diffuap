```python
import torch
import torchvision.transforms as transforms

def normalize(input_tensor):
    mean = input_tensor.mean(dim=(0, 2, 3))
    std = input_tensor.std(dim=(0, 2, 3))
    normalize = transforms.Normalize(mean, std)
    return normalize(input_tensor) ,mean ,std

def unnormalize(input_tensor, mean, std):
    unnormalize = transforms.Normalize(
    mean=-mean/std,
    std=1/std
)
    return unnormalize(input_tensor)
```

```python
import numpy as np
latent = np.load("latent.npy")
print(latent.shape)
import torch
input_tensor = torch.tensor(latent)
output, mean,std = normalize(input_tensor)
unoutput = unnormalize(output,mean,std)
```
